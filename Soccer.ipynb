```python
# import modules
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# sklearn utilities
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

# sklearn models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
```

```python
path_events = os.getcwd() + '\events.csv'
events = pd.read_csv(path_events)
path_ginf = os.getcwd() + '\ginf.csv'
ginf = pd.read_csv(path_ginf)

events.dropna()
ginf.dropna()

events = events.drop(['id_odsp', 'id_event', 'text', 'event_team', 'opponent', 'player', 'player2', 'player_in', 'player_out'], axis = 'columns')
ginf = ginf.drop(['id_odsp', 'link_odsp'], axis = 'columns')
events = events.fillna(0)

game = []
prev = 0
count = 0
for i in events['time']:
    if prev > i:
        count = count + 1
    game.append(count)
    prev = i

events['Game'] = game
events
```

```python
event_X = events
event_y = events.iloc[0:,10]
x_train, x_test, y_train, y_test = train_test_split(event_X, event_y, test_size = .25, random_state = 0)
```

```python
RFC = RandomForestClassifier(n_estimators = 10)
RFC.fit(x_train, y_train)
print(RFC.score(x_test, y_test))
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
predict_X = lr.predict(x_test)
```
